```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

KNN MODEL
The KNN model classifies a new data point based on its similarity to the k-nearest neighbors in the training dataset. The formula for the KNN model is:

y = f(x) = argmax(k=1...K, d(x, x_k))

where y is the predicted class, x is the new data point, d(x, x_k) is the distance between x and the kth neighbor in the training dataset, and argmax is the function that returns the value that maximizes the given function.

This formula represents the KNN model, which uses the distances to the nearest neighbors in the training dataset to classify new data points.

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy: ", model.score(X_test, y_test))
```
